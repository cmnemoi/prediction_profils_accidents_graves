```python
import pandas as pd

# Chemin du fichier Parquet
file_path = "/content/dataset_clean.parquet"

# Lecture du fichier Parquet
dataset_clean = pd.read_parquet(file_path)

display(dataset_clean.describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Convert columns with time values to numerical representation
# Assuming 'heure_accident' is the column containing time values like '07:32'
# You might need to adjust the column name if it's different in your dataset
#dataset_clean['heure_accident'] = pd.to_datetime(dataset_clean['heure_accident'], format='%H:%M').dt.hour * 60 + pd.to_datetime(dataset_clean['heure_accident'], format='%H:%M').dt.minute


# Select only numerical features for correlation analysis
numerical_features = dataset_clean.select_dtypes(include=np.number)
correlation_matrix = numerical_features.corr()

# Masquer la moitié supérieure de la matrice
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Création de la figure avec des dimensions ajustées
plt.figure(figsize=(14, 12))

# Heatmap avec des ajustements
sns.heatmap(
    correlation_matrix,
    mask=mask,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    annot=True,
    annot_kws={"size": 8},  # Taille des annotations
    linewidths=0.8,  # Espacement entre les cases
    square=True  # Forcer des cases carrées
)

# Rotation des étiquettes pour améliorer la lisibilité
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)

# Titre du graphique
plt.title("Heatmap des corrélations entre les variables", fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```

 Analyse par genre : Proportion des hommes et des femmes dans la mortalité routière

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ajouter une colonne pour le genre
dataset_clean['genre'] = dataset_clean['sexe_usager'].map({1: 'Homme', 2: 'Femme'})

# Filtrer les décès uniquement et sur l'année 2022
deaths_2022 = dataset_clean[(dataset_clean['gravite_blessure'] == 2) & (dataset_clean['annee_accident'] == 2022)]

# Calculer les proportions par genre
proportion_genre_2022 = deaths_2022['genre'].value_counts(normalize=True) * 100
print(proportion_genre_2022)

# Visualisation avec affichage des pourcentages
plt.figure(figsize=(8, 6))
bars = plt.bar(proportion_genre_2022.index, proportion_genre_2022.values, color=['blue', 'pink'])
plt.title('Proportion des décès par genre en 2022', fontsize=14)
plt.ylabel('Pourcentage', fontsize=12)
plt.xlabel('Genre', fontsize=12)

# Ajouter les pourcentages sur les barres
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}%', ha='center', va='bottom', fontsize=10)

plt.show()
```

Analyse par âge : Distribution des tués par classe d'âge

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ajouter une colonne pour l'âge
dataset_clean['age'] = 2023 - dataset_clean['annee_naissance']

# Catégoriser les âges
bins = [0, 14, 17, 24, 34, 64, 74, 120]
labels = ['<15 ans', '14-17 ans', '18-24 ans', '25-34 ans', '35-64 ans', '65-74 ans', '75 ans et plus']
dataset_clean['classe_age'] = pd.cut(dataset_clean['age'], bins=bins, labels=labels, right=False)

# Filtrer les décès uniquement pour 2022 et 2023
deaths_2022 = dataset_clean[(dataset_clean['gravite_blessure'] == 2) & (dataset_clean['annee_accident'] == 2022)]
deaths_2023 = dataset_clean[(dataset_clean['gravite_blessure'] == 2) & (dataset_clean['annee_accident'] == 2023)]

# Comptage par classe d'âge pour chaque année
age_distribution_2022 = deaths_2022['classe_age'].value_counts().sort_index()
age_distribution_2023 = deaths_2023['classe_age'].value_counts().sort_index()

# Afficher les distributions
print("Répartition des décès par classe d'âge en 2022 :")
print(age_distribution_2022)
print("\nRépartition des décès par classe d'âge en 2023 :")
print(age_distribution_2023)

# Visualisation
plt.figure(figsize=(12, 6))
bar_width = 0.35
x = range(len(age_distribution_2022))

bars_2022 = plt.bar(x, age_distribution_2022, width=bar_width, color='blue', label='2022')
bars_2023 = plt.bar([i + bar_width for i in x], age_distribution_2023, width=bar_width, color='orange', label='2023')

# Ajouter les étiquettes avec les valeurs exactes
for bar in bars_2022:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1, int(bar.get_height()), ha='center', fontsize=10)
for bar in bars_2023:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1, int(bar.get_height()), ha='center', fontsize=10)

# Ajouter les labels et légendes
plt.title('Répartition des décès par classe d\'âge (2022 vs 2023)', fontsize=14)
plt.ylabel('Nombre de décès', fontsize=12)
plt.xlabel('Classe d\'âge', fontsize=12)
plt.xticks([i + bar_width / 2 for i in x], labels)
plt.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrer les données pour les années 2022 et 2023
data_2022 = dataset_clean[dataset_clean['annee_accident'] == 2022]
data_2023 = dataset_clean[dataset_clean['annee_accident'] == 2023]

# 1. Calculer le nombre de tués
deaths_2022 = data_2022[data_2022['gravite_blessure'] == 2].shape[0]
deaths_2023 = data_2023[data_2023['gravite_blessure'] == 2].shape[0]

# Calculer la variation en pourcentage
deaths_variation = ((deaths_2023 - deaths_2022) / deaths_2022) * 100

# 2. Calculer le nombre de blessés graves
serious_injuries_2022 = data_2022[data_2022['gravite_blessure'] == 3].shape[0]
serious_injuries_2023 = data_2023[data_2023['gravite_blessure'] == 3].shape[0]

# Calculer la variation en pourcentage
serious_injuries_variation = ((serious_injuries_2023 - serious_injuries_2022) / serious_injuries_2022) * 100

# 3. Calculer le nombre total de blessés toutes gravités
total_injuries_2022 = data_2022.shape[0]
total_injuries_2023 = data_2023.shape[0]

# Calculer la variation en pourcentage
total_injuries_variation = ((total_injuries_2023 - total_injuries_2022) / total_injuries_2022) * 100

# Afficher les résultats
print(f"Tués en 2022 : {deaths_2022}, en 2023 : {deaths_2023}, variation : {deaths_variation:.1f}%")
print(f"Blessés graves en 2022 : {serious_injuries_2022}, en 2023 : {serious_injuries_2023}, variation : {serious_injuries_variation:.1f}%")
print(f"Blessés toutes gravités en 2022 : {total_injuries_2022}, en 2023 : {total_injuries_2023}, variation : {total_injuries_variation:.1f}%")

# 4. Visualisation
labels = ['Tués', 'Blessés graves', 'Blessés toutes gravités']
values_2022 = [deaths_2022, serious_injuries_2022, total_injuries_2022]
values_2023 = [deaths_2023, serious_injuries_2023, total_injuries_2023]

x = range(len(labels))
bar_width = 0.35

plt.figure(figsize=(10, 6))
bars_2022 = plt.bar(x, values_2022, width=bar_width, color='blue', label='2022')
bars_2023 = plt.bar([i + bar_width for i in x], values_2023, width=bar_width, color='orange', label='2023')

# Ajouter des étiquettes avec les valeurs
for bar in bars_2022:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 500, int(bar.get_height()), ha='center', fontsize=10)
for bar in bars_2023:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 500, int(bar.get_height()), ha='center', fontsize=10)

# Ajouter les labels et légendes
plt.title('Comparaison de la mortalité et des blessés (2022 vs 2023)', fontsize=14)
plt.ylabel('Nombre de cas', fontsize=12)
plt.xlabel('Catégories', fontsize=12)
plt.xticks([i + bar_width / 2 for i in x], labels)
plt.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```
